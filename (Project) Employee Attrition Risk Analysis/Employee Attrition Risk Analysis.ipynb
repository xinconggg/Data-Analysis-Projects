## Feature Engineering
Create new features or transform existing ones to better capture the patterns that influence attrition. This is done to significantly improve model accuracy and interpretability.

#### 1) Aggregate Categorical Features
Group job roles and departments to reduce feature sparsity:
- **MangementRole:** `Manager`, `Manufacturing Director` and `Research Director`
- **TechnicalRole:** `Laboratory Technician` and `Research Scientist`
- **CustomerFacingRole:** `Sales Executive` and `Sales Representative`
- **SupportRole:** `Human Resources`

```python
# Technical Roles: Laboratory Technician, Research Director, Research Scientist, Manufacturing Director
df['TechnicalRole'] = df[['JobRole_Laboratory Technician',
                          'JobRole_Research Director',
                          'JobRole_Research Scientist',
                          'JobRole_Manufacturing Director']].sum(axis=1)

# Management Roles: Manager, Manufacturing Director, Research Director
df['ManagementRole'] = df[['JobRole_Manager',
                           'JobRole_Manufacturing Director',
                           'JobRole_Research Director']].sum(axis=1)

# Customer-Facing Roles: Sales Executive, Sales Representative
df['CustomerFacingRole'] = df[['JobRole_Sales Executive',
                               'JobRole_Sales Representative']].sum(axis=1)

# Support Roles: Human Resources
df['SupportRole'] = df[['JobRole_Human Resources']].sum(axis=1)
```

#### 2) Ratio-Based Features (Experience and Growth)
To capture an employee's progression and stability within the company:
- **Career Stability:**
  - Years at company relative to total working years
  - High value indicates loyalty to company
- **Current Role Stability:**
  - Years in current role relative to total working years
  - Captures potential for growth within the same role
- **Promotion Frequency:**
  - Years since last promotion relative to years at company
  - Frequent promotions might indicate a reduction in attrition risk

```python
df['CareerStability'] = df['YearsAtCompany'] / (df['TotalWorkingYears']+0.000001) # +0.000001 to avoid division by 0
df['RoleStability'] = df['YearsInCurrentRole'] / (df['TotalWorkingYears']+0.000001)
df['PromotionFrequency'] = (df['YearsAtCompany'] - df['YearsSinceLastPromotion']) / (df['YearsAtCompany']+0.000001)
```

#### 3) Overall Satisfaction
To create a more holistic view of employee satisfaction, which is often a critical factor in attrition decisions.
Calculated by taking the mean of `JoJobSatisfaction`, `EnvironmentSatisfaction`, `RelationshipSatisfaction` and `WorkLifeBalance`.


```python
df['OverallSatisfaction'] = (df['JobSatisfaction'] + df['EnvironmentSatisfaction'] + df['RelationshipSatisfaction'] + df['WorkLifeBalance']) / 4
```

#### 4) Financial Stress Indicator
To assess financial satisfaction:
- **IncomeToHikeRatio:** Lower ratios might suggest dissatisfaction despite pay hikes, potentially signaling risk
- **IncomePerYear:** To capture financial status without relying on monthly income alone


```python
df['IncomeToHikeRatio'] = df['MonthlyIncome'] / (df['PercentSalaryHike'])
df['IncomePerYear'] = df['MonthlyIncome'] * 12
```

#### 5) Work-Life Balance and Commitment
Assess the work-life balance and commitment to the company:
- **OvertimeImpact:** Overworked employees who work overtime frequently, with poor work-life balance are more likely to leave the company
- **YearsPerRole:** To identify if an employee have been stucked in their current role for too long

```python
df['OvertimeImpact'] = df['OverTime_Yes'] * df['WorkLifeBalance']
df['YearsPerRole'] = df['YearsAtCompany'] / (df['JobLevel'])
```

#### 6) High-Risk Group
Create features for high-risk groups based on known high-attrition factors:
- **HighRiskLowSatisfaction:** Employees who are both unhappy with their job and uncomfortable in their work environment are much more likely to seek other opportunities
- **LongTimeNoPromotion:** Employees who haven’t received a promotion in more than 5 years might feel undervalued or stuck, leading to disengagement
- **FrequentJobChanger:** A count of more than 4 companies suggests a pattern of low organizational loyalty or a tendency to move for better opportunities

```python
df['HighRiskLowSatisfaction'] = (df['JobSatisfaction'] <= 2) & (df['EnvironmentSatisfaction'] <= 2)
df['LongTimeNoPromotion'] = df['YearsSinceLastPromotion'] > 5
df['FrequentJobChanger'] = df['NumCompaniesWorked'] > 4
```

### Correlation Heatmap
Visualize the correlation (ρ) between different numerical variables in a dataset. It shows how strongly variables are related to each other. If 2 features are highly correlated, one of them may be redundant, and dropping one might help reduce multicollinearity (which can cause issues in some models).

| \|ρ\| Value Range    | Correlation Strength         | Interpretation                          |
|---------------------|-----------------------------|------------------------------------------|
| 0.8 ≤ \|ρ\| ≤ 1.0    | Very Strong Correlation      | Strong, almost perfect relationship      |
| 0.6 ≤ \|ρ\| < 0.8    | Strong Correlation           | Significant relationship, but not perfect|
| 0.4 ≤ \|ρ\| < 0.6    | Moderate Correlation         | Moderate relationship                    |
| 0.2 ≤ \|ρ\| < 0.4    | Weak Correlation             | Weak, but noticeable relationship        |
| 0.0 ≤ \|ρ\| < 0.2    | Very Weak Correlation        | Very weak, close to no relationship      |

### Multicollinearity Check using Variance Inflation Factor (VIF)
The following code calculates the VIF, which is used to detect multicollinearity among numerical features. If two or more features are highly correlated, it can lead to multicollinearity, which can make the model unstable.

| VIF Value Range        | Multicollinearity Level      | Interpretation                                           |
|------------------------|------------------------------|----------------------------------------------------------|
| 1                      | No multicollinearity         | Ideal (No issues, perfect scenario)                      |
| 1 - 5                  | Moderate multicollinearity   | Acceptable, but monitor the variables for changes        |
| 5 - 10                 | High multicollinearity       | Potential problem, consider removing or combining features |
| > 10                   | Severe multicollinearity     | Strongly consider removing or transforming variables     |





```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.preprocessing import StandardScaler

# Extract the numerical features
X = df[numerical_features]

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Create a DataFrame to hold the VIF values
vif_data = pd.DataFrame()
vif_data['Feature'] = numerical_features

# Calculate VIF for each feature
vif_data['VIF'] = [variance_inflation_factor(X_scaled, i) for i in range(X_scaled.shape[1])]

# Print the VIF values
vif_data
```

#### Filtering and Removal of Variables with High VIF (≥ 5)

```python
# Filter for high VIF values (e.g., VIF >= 5)
high_vif_features = vif_data[vif_data['VIF'] >= 5]
high_vif_features
```

```python
# Get the list of high VIF feature names
high_vif_feature_names = high_vif_features['Feature'].tolist()

df.drop(columns=high_vif_feature_names, inplace=True)
```

### Feature Selection
Helps to reduce dimensionality, improve model performance and reduce overfitting.

#### 1) Random Forest
To identify the most important features based on their contribution to the model's decision. Features with low importance will be dropped or deprioritized.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder

# Splitting of data
X = df.drop(columns=['Attrition'])
y = df['Attrition']

# Encode non-numeric columns
label_encoder = LabelEncoder()
for col in X.select_dtypes(include=['object']).columns:
    X[col] = label_encoder.fit_transform(X[col])

# Train a simple RandomForest model to estimate feature importance
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X, y)

# Plot feature importance
importances = rf.feature_importances_
features = X.columns
indices = importances.argsort()

plt.figure(figsize=(15, 10))
plt.title("Feature Importances")
plt.barh(range(len(indices)), importances[indices], align="center")
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.show()
```

Judging from the feature importance plot, most of the features that has <0.02 feature importance and should be removed area:
- `Over18`
- `JobRole`, `SupportRole`, `TechnicalRole`, `CustomerFacingRole`, `ManagementRole`
- `Department`
- `EducationField`
- `LongTimeNoPromotion`
- `PerformanceRating`
- `MaritialStatus`
- `Gender`
- `FrequentJobChanger`
- `HighRiskLowSatisfaction`
- `BusinessTravel`

```python
# List of base of low-importance one-hot encoded features to remove
base_features_to_remove = ['JobRole', 'EducationField', 'Department', 'MaritalStatus', 'Gender', 'BusinessTravel']

# List of un-encoded low importance features
low_importance_features = ['Over18', 'SupportRole', 'TechnicalRole', 'CustomerFacingRole', 'ManagementRole', 'LongTimeNoPromotion', 'PerformanceRating', 'FrequentJobChanger', 'HighRiskLowSatisfaction']

# Combine all features to remove
all_features_to_remove = low_importance_features.copy()

# Add all one-hot encoded versions of the base features
for base_feature in base_features_to_remove:
    all_features_to_remove.extend([col for col in X.columns if col.startswith(base_feature + '_')])

# Drop the identified features
X.drop(columns=all_features_to_remove, inplace=True)
```

#### 2) Recursive Feature Elimination (RFE)
To further reduce the number of features based on the model's performance and select only the top 15 features.

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

# Using Logistic Regression for RFE
log_reg = LogisticRegression(max_iter=1000, random_state=42)
rfe = RFE(log_reg, n_features_to_select=15)
rfe.fit(X, y)

# Selected features
selected_features = X.columns[rfe.support_]
print("Selected Features:", selected_features)
```

#### 1) Logistic Regression (Baseline Model)
- Since Logistic Regression is a simple model, it will be used to get an initial baseline

### Evaluation and Comparison of Models
A good way to evaluate classification models is using:
- **Precision**:
  - $\text{Precision} = \frac{\text{True Positives}}{\text{True Positives } + \text{ False Positives}}$
  - Proportion of **positive predictions** that are actually correct
- **Recall**:
  - $\text{Recall} = \frac{\text{True Positives}}{\text{True Positives } + \text{ False Negatives}}$
  - Proportion of **actual positives** that are actually correct
- **F1-score**:
  - $\text{F1-Score} = 2*\frac{\text{Precision * Recall}}{\text{Precision } + \text{ Recall}}$
  - Harmonic mean of precision and recall
  - A higher score indicates better balance between precision and recall
- **ROC-AUC (Receiver Operating Characteristic-Area Under Curve)**:
  - $\text{ROC-AUC} = \int_0^1 \text{TPR} \,  d(\text{FPR})$
  - Evaluates the overall **classification capability** at all threshold
  - **Interpretation:**
    - AUC = 0.5: The model is no better than random guessing
    - AUC > 0.8: Considered a good model
    - AUC = 1: The model has perfect discrimination between classes



```python
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt

# Calculate ROC-AUC for each model
models = {
    "Logistic Regression": log_reg,
    "Decision Tree": dt,
    "Random Forest": rf,
    "XGBoost": xgb
}

plt.figure(figsize=(10, 8))

for name, model in models.items():
    # ROC-AUC Score
    if hasattr(model, "predict_proba"):
        roc_auc = roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])
        print(f"{name} ROC-AUC: {roc_auc:.4f}")

        # ROC Curve
        fpr, tpr, _ = roc_curve(y_test, model.predict_proba(X_test)[:, 1])
        plt.plot(fpr, tpr, label=f"{name} (AUC = {roc_auc:.2f})")
    else:
        print(f"{name} does not support probability prediction and was skipped.")

# Add diagonal reference line
plt.plot([0, 1], [0, 1], color='navy', linestyle='--', label="Random Classifier (AUC = 0.50)")

# Add plot details
plt.title('ROC Curve Comparison')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.show()
```

| **Decision Tree**        | 0.5958            | Poor discriminatory ability, only slightly better than random guessing. Likely overfitting or failing to capture meaningful patterns. |

### Hyperparameter Tuning
Fine-tune the model's hyperparameter to improve the models' performance. Based off the ROC-AUC score, **decision tree** can be removed due to its poor ROC-AUC score, likely caused by being too sensitive to noise, leading to high variance.
#### 1) Hyperparameter Tuning and Re-fitting of Logistic Regression

```python
# Hyperparameter Tuning
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

# Define the parameter grid with only compatible combinations
param_grid_lr = [
    {'C': [0.01, 0.1, 1, 10, 100], 'penalty': ['l1'], 'solver': ['liblinear', 'saga']},
    {'C': [0.01, 0.1, 1, 10, 100], 'penalty': ['l2'], 'solver': ['lbfgs', 'liblinear', 'saga']},
    {'C': [0.01, 0.1, 1, 10, 100], 'penalty': ['elasticnet'], 'solver': ['saga'], 'l1_ratio': [0.1, 0.5, 0.9]},
    {'C': [0.01, 0.1, 1, 10, 100], 'penalty': [None], 'solver': ['lbfgs', 'saga']}
]

# Run GridSearchCV
grid_search_lr = GridSearchCV(
    LogisticRegression(max_iter=1000),
    param_grid=param_grid_lr,
    scoring='accuracy',
    cv=5,
    n_jobs=-1,
    verbose=2
)

grid_search_lr.fit(X_train, y_train)

print("Best Parameters for Logistic Regression:", grid_search_lr.best_params_)
print("Best Cross-Validation Score:", grid_search_lr.best_score_)
```

### Class Imbalance Handling
#### Synthetic Minority Oversampling Technique (SMOTE)
A popular oversampling technique used to address class imbalance. It generates synthetic samples for the minority class, making the dataset more balanced, allowing the model to learn better from the under-represented class.

```python
from imblearn.over_sampling import SMOTE

# Apply SMOTE to balance the classes
smote = SMOTE(sampling_strategy='minority', random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# Logistic Regression
lr_smote = LogisticRegression(**best_lf_params)
lr_smote.fit(X_train_smote, y_train_smote)
lr_smote_cv_scores = cross_val_score(lr_smote, X_train_smote, y_train_smote, cv=10)
print("Logistic Regression (SMOTE)")
print(classification_report(y_test, lr_smote.predict(X_test)))
print("\nAverage Cross-Validation Score:", lr_smote_cv_scores.mean())

# Train Random Forest
rf_smote = RandomForestClassifier(**best_rf_params)
rf_smote.fit(X_train_smote, y_train_smote)
rf_smote_cv_scores = cross_val_score(rf_smote, X_train_smote, y_train_smote, cv=10)
print("\nRandom Forest (SMOTE)")
print(classification_report(y_test, rf_smote.predict(X_test)))
print("\nAverage Cross-Validation Score:", rf_smote_cv_scores.mean())

# Train XGBoost
xgb_smote = XGBClassifier(**best_xgb_params)
xgb_smote.fit(X_train_smote, y_train_smote)
xgb_smote_cv_scores = cross_val_score(xgb_smote, X_train_smote, y_train_smote, cv=10)
print("\nXGBoost (SMOTE)")
print(classification_report(y_test, xgb_smote.predict(X_test)))
print("\nAverage Cross-Validation Score:", xgb_smote_cv_scores.mean())
```
